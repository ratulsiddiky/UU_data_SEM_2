## Exercise 2

The casino wheel is divided into ten sections labeled 1 through 10. Each section has a payout based on the number it represents: section 1 pays 1-to-1, section 2 pays 2-to-1, and so on up to section 10, which pays 10-to-1.

The probability of the ball landing on section number $n > 10$ is $\frac{1}{2n^2}$, while the probability of landing on section 10 is $0.2251$.

### Informal Expected Value Calculation

To determine the best bet and optimal betting strategy, we calculate the expected value (EV) for each section of the wheel.

- **For sections 1 through 10:**
    - Section 1: Payout = 1, Probability = 1 (certain event), EV = 1 * 1 = 1
    - Section 2: Payout = 2, Probability = $\frac{1}{4}$ (given), EV = 2 * $\frac{1}{4}$ = $\frac{1}{2}$
    - Section 3: Payout = 3, Probability = $\frac{1}{18}$ (given), EV = 3 * $\frac{1}{18}$
    - ...
    - Section 10: Payout = 10, Probability = 0.2251 (given), EV = 10 * 0.2251 = 2.251

- **For sections with numbers greater than 10:**
    - Probability = $\frac{1}{2n^2}$ (given)
    - Since the payout is the same as the section number, the EV for each section is simply the probability multiplied by the section number.

### Best Bet and Optimal Strategy

Comparing the expected values:
- For sections 1 through 10, the EV ranges from 1 to 2.251.
- For sections with numbers greater than 10, the EV decreases as the section number increases due to the decreasing probability.

The best bet is on section 10, as it has the highest expected value. Therefore, the optimal betting strategy would be to consistently bet on section 10, as it offers the highest average winnings.

## Exercise 5:
Let's denote the following events:
- $ P(H)=1/3 $: The grand prize is behind Door A.
- $P(H^c)=2/3$: A Goat is behind door A.



The host then opens Door B, revealing a goat. Now, Benny needs to decide whether to stick with Door A or switch to Door C.

#### Stick with Door A:
If Benny sticks with Door A, he wins if the grand prize is behind Door A. The probability of winning does not change in this case and it is simply $ P(H) = \frac{1}{3} $.

#### Switch to Door C:
If Benny switches to Door C, we are able to apply Bayes' Theorem. Let us compute the probability of the prize being behind door A given that behind door B is a goat (Event $E$).

$$P(H|E)=\frac{P(E|H)P(H)}{P(E)}=\frac{P(E|H)P(H)}{P(E|H)P(H)+P(E|H^c)P(H^c)}$$

$P(H)=1/3$. $P(H^c)=2/3$. $P(E|H)$ is the probability that Benny shows a door with a goat behind it, given that there is a car behind door A. Since Monty always shows a door with a goat, this is equal to 1. $P(E|H^c)$ is the probability that Benny shows the goat, given that there is a goat behind door A. Again, since Monty always shows a door with a goat, this is equal to 1.

Therefore,
$$P(H|E)=\frac{P(E|H)P(H)}{P(E)}=\frac{P(E|H)P(H)}{P(E|H)P(H)+P(E|H^c)P(H^c)}=\frac{1/3}{1/3*2/3}=1/3.$$

#### Conclusion:
The probability that the car is behind door A is completely unchanged by the evidence. However, since the car can only either be behind door A or behind the door Benny didn't reveal, the probability of the car behind the door that is not revealed is $1-P(H|E)=2/3$. Therefore, switching is twice as likely to get you the car as staying.
