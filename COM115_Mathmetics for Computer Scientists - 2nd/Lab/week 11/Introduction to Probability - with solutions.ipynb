# Week 11: Introduction to Probability


Probability, one of the main modern fields in mathematics and statistics, serves as a fundamental tool for understanding uncertainty and quantifying randomness. It is broadly used in different fileds, from pure mathematics and physics to economics or social sciences, and   provides a rigorous framework to understand uncertain events and making informed decisions.

At its core, we could say that probability theory enables to analyze the likelihood of various outcomes in a given situation, whether it be predicting the outcome of a randomized experiment, assessing risk in financial markets, or interpreting data in scientific research. Before we move on to practical applications, it is important to define the basis of this topic.


## 11.1. Definitions

### Experiment

In this context, an experiment is any process or procedure that results in one or more outcomes. It is a controlled action (or observation) that can produce a set of well-defined outcomes It can be as simple as flipping a coin or rolling a die, or as complex as conducting a scientific study.

Examples of experiments include tossing a coin, drawing a card from a deck, conducting a survey, or observing the outcome of a medical treatment.

### Sample Space

The sample space, denoted by  $S$, is the set of all possible outcomes of an experiment. It includes every conceivable outcome that could happen, with no repetition and no omissions.

The sample space can be finite, countably infinite, or uncountably infinite, depending on the nature of the experiment.

For example, when rolling a fair six-sided die, the sample space  $S$ consists of the numbers $\{1, 2, 3, 4, 5, 6\}$. Each number represents a possible outcome of the experiment.

In more complex experiments, the sample space may involve combinations of outcomes. For instance, in flipping two coins, the sample space includes $\{HH, HT, TH, TT\}$, where H represents heads and T represents tails.

### Events

An event is a subset of the sample space, representing a specific outcome or a collection of outcomes in an experiment. In other words, an event is a particular situation that we are interested in observing or analyzing.

Events are denoted using capital letters, such as  $A$, $B$, etc. They can range from simple events, representing a single outcome, to complex events, consisting of multiple outcomes.

For example:
- In the experiment of rolling a six-sided die, the event of rolling an even number can be denoted as $A = \{2, 4, 6\} $.
- In flipping two coins, the event of getting at least one head can be denoted as $ B = \{HH, HT, TH\} $.

#### Complementary Event:

The complementary event of an event $ A $ is the event consisting of all outcomes that are not in $ A $. In other words, it represents the occurrence of the opposite or negation of event $ A $. The complementary event is denoted by $ A^c $ (as in set theory). Mathematically, it can be defined as:

$$A^c = \{ \text{outcomes not in } A \} $$

#### Union of Events:

The union of two events $ A $ and $ B $, denoted by $ A \cup B $, is the event that occurs if at least one of the events $ A $ or $ B $ occurs. In other words, it represents the combination of outcomes belonging to either event $ A $ or event $ B $, or both. Mathematically, it can be defined as:

$$ A \cup B = \{ \text{outcomes in } A \text{ or } B \text{ or both} \} $$

#### Intersection of Events:

The intersection of two events $ A $ and $ B $, denoted by $ A \cap B $, is the event that occurs if both events $ A $ and $ B $ occur simultaneously. In other words, it represents the set of outcomes that are common to both events $ A $ and $ B $. Mathematically, it can be defined as:

$$ A \cap B = \{ \text{outcomes in both } A \text{ and } B \} $$


### Probability

Probability is number in $[0,1]$ representing the likelihood of an event happening. It quantifies uncertainty and provides a way to predict the likelihood of different outcomes. The probability of an event $ A $, denoted by $ P(A)$.

Probabilities are typically calculated based on the relative frequency of an event occurring in repeated trials of an experiment, under similar conditions. However, they can also be determined using theoretical models or logical reasoning, depending on the nature of the problem.

#### Independent Events:

Two events $ A $ and $ B $ are independent if the occurrence of one event does not affect the occurrence of the other. In other words, the probability of event $ A $ occurring is the same regardless of whether event $ B $ has occurred, and vice versa. Mathematically, two events $ A $ and $ B $ are independent if:

$$P(A \cap B) = P(A)P(B) $$

This means that the probability of both events happening is equal to the product of their individual probabilities.


## 11.2. Kolmogorov Axioms

The Kolmogorov axioms are the fundamental principles in probability theory. They were proposed by the Russian mathematician Andrey Kolmogorov (1903-1987). These axioms provide the basis from where probability theory is built. 

1. **Non-negativity Axiom**: 
   - The probability of any event $ A $ is a non-negative real number.
   - Mathematically, for any event $A$, $P(A) \geq 0$.

2. **Normalization Axiom**:
   - The probability of the entire sample space $ S $ is equal to 1.
   - Mathematically, $P(S) = 1 $.

3. **Additivity Axiom** :
   - If $ A_1, A_2, A_3, \ldots \in S$ are mutually exclusive events (i.e., they have no outcomes in common), then the probability of the union of these events is equal to the sum of their individual probabilities.
   - Mathematically, given $ A_1, A_2, A_3, \ldots $ such that $A_i\cap A_j=\varnothing $ for every $i\neq j$, then
     $$
     P(\bigcup_{i=1}^{\infty} A_i) = \sum_{i=1}^{\infty}P(A_i)
     $$

These axioms serve as the building blocks of probability theory, "ensuring" that probabilities behave in a consistent and logical manner. They provide a formal framework for defining probability measures on various spaces, allowing for the rigorous analysis of random phenomena in diverse fields.


## 11.3. Example: Flipping a (Fair) Coin

Suppose we have a fair coin, which means that it has an equal chance of landing on heads (H) or tails (T) when flipped. We want to calculate the probability of certain events related to the outcomes of flipping this coin. My sample space is the set of all possible outcomes $S=\{H,T\}$.

#### Event 1: Probability of Getting Heads (H)

The probability of getting heads $P(\text{H}) $ can be computed using a relative frequency approach. Since the coin is fair, there is one outcome corresponding to heads out of the two possible outcomes (heads or tails). Therefore, we have:

$$ P(\text{H}) = \frac{\text{Number of favorable outcomes}}{\text{Total number of outcomes}} = \frac{1}{2} $$

So, the probability of getting heads when flipping a fair coin is $ \frac{1}{2} $ or $0.5$.

#### Event 2: Probability of Getting Tails (T)

Similarly, the probability of getting tails $ P(\text{T}) $ is also $ \frac{1}{2} $ since there are two equally likely outcomes (heads or tails), and each outcome has the same probability.

$$ P(\text{T}) = \frac{\text{Number of favorable outcomes}}{\text{Total number of outcomes}} = \frac{1}{2} $$

#### Event 3: Probability of Getting Either Heads or Tails (H or T)

The probability of getting either heads or tails $ P(\text{H or T}) $ can be calculated using the Kolmogorov additivity axiom. Since the events "getting heads" and "getting tails" are mutually exclusive (they cannot happen simultaneously), we can sum their individual probabilities:

$$ P(\text{H or T})=P(\{H, T\}) = P(\text{H}) + P(\text{T}) = \frac{1}{2} + \frac{1}{2} = 1 $$

So, the probability of getting either heads or tails when flipping a fair coin is 1, as expected, because one of these outcomes is certain to occur.

## 11.6. Conditional Probability

Conditional probability is a concept in probability theory that measures the likelihood of an event occurring given that another event has already happened.

### Definition:

The conditional probability of an event $ A $ given that an event $ B $ has occurred is denoted by $ P(A|B) $, read as "the probability of $ A $ given $ B $. It is defined as the ratio of the probability of the intersection of events $ A$ and $ B $ to the probability of event $ B $, provided that $P(B) > 0 $. Mathematically, it can be expressed as:

$$ P(A|B) = \frac{P(A \cap B)}{P(B)}. $$


### Interpretation:

Conditional probability allows us to update our knowledge about the likelihood of an event occurring based on new information provided by the occurrence of another event. It helps in making predictions and decisions in situations where outcomes depend on certain conditions or constraints.

#### Example:

Consider the experiment of drawing two cards successively from a standard deck of 52 playing cards without replacement. What is the probability of getting two red cards? 

Now, let us supposed that we draw the first card and it was red. Let $ A $ be the event of drawing a red card on the first draw, and let $B$ be the event of drawing a red card on the second draw. The conditional probability of drawing a red card on the second draw given that a red card was drawn on the first draw can be calculated as:

$$P(B|A) = \frac{P(A \cap B)}{P(A)} $$

This example illustrates how conditional probability helps in assessing the likelihood when the information we have is updated.

## 11.8. Bayes Theory

### Introduction to Bayes' Theorem

Bayes' Theorem is a fundamental concept in probability theory named after the Reverend Thomas Bayes, an 18th-century mathematician. It provides a mathematical framework for updating beliefs or predictions based on new evidence or information.

Bayes' Theorem is a method for calculating conditional probabilities. It formalizes the process of updating our beliefs about the probability of an event as we receive new information.

#### Bayes' Theorem:

Bayes' Theorem is often expressed as:

$$ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} $$

Where:
- $ P(A|B) $ is the probability of event $ A $ occurring given that event $ B $ has occurred.
- $ P(B|A) $ is the probability of event $ B $ occurring given that event $ A $ has occurred.
- $ P(A) $ and $ P(B) $ are the probabilities of events $ A $ and $ B $ occurring independently of each other.

#### Interpretation:

Bayes' Theorem provides a systematic way to update our beliefs or predictions about the probability of an event based on new evidence or information. It allows us to incorporate prior knowledge (represented by $ P(A) $) and new observations (represented by $ P(B|A) $) to update our understanding of the probability of the event of interest.

#### Applications:

Bayes' Theorem has wide-ranging applications across various fields, including:
- **Medical Diagnosis**: Updating the probability of a disease given the results of diagnostic tests.
- **Machine Learning**: Bayesian inference is used in probabilistic models to make predictions and classify data.
- **Finance**: Assessing the likelihood of financial events such as market movements or default probabilities.
- **Natural Language Processing**: Bayesian methods are employed in language modeling and text classification tasks.

#### Bayesian vs. Frequentist Approaches:

Bayesian statistics differs from frequentist statistics in its treatment of uncertainty. While frequentist statistics relies on repeated sampling and probability distributions of observed data, Bayesian statistics incorporates prior beliefs about the parameters being estimated and updates these beliefs as new data becomes available.




### Law of Total Probability

The Law of Total Probability is a fundamental concept in probability theory that allows us to express the probability of an event in terms of conditional probabilities. Let $ B_1, B_2, ..., B_n $ be a partition of the sample space $ S $, then for any event $ A $ in $ S $, we have:

$$ P(A) = \sum_{i=1}^{n} P(A|B_i) P(B_i) $$

Where:
- $ P(A|B_i) $ is the conditional probability of event $ A $ occurring given that event $ B_i $ has occurred.
- $ P(B_i) $ is the probability of event $ B_i $ occurring.

The Law of Total Probability is often used to compute probabilities of events by considering all possible scenarios or partitions of the sample space.

#### 1. Drawing Two Balls:

a) Probability of drawing a red ball followed by a blue ball $ P(R \cap B): $

To compute this probability, we need to consider the probability of drawing a red ball on the first draw $$ P(R) $$ and then drawing a blue ball on the second draw, given that a red ball was already drawn $$ P(B|R). $$ Since the balls are drawn without replacement, the events are not independent.

$$ P(R \cap B) = P(R) \times P(B|R) $$

$$ P(R \cap B) = 0.4 \times \frac{3}{9} = \frac{2}{15} $$

b) Probability of drawing two balls of the same color $$ P(R \cap R) + P(B \cap B) + P(W \cap W): $$

Since the balls are drawn without replacement, we have to consider the probability of drawing the same color ball twice for each color.

$$ P(R \cap R) = P(R) \times P(R|R) $$
$$ P(B \cap B) = P(B) \times P(B|B) $$
$$ P(W \cap W) = P(W) \times P(W|W) $$

$$ P(R \cap R) = 0.4 \times \frac{3}{9} = \frac{2}{15} $$
