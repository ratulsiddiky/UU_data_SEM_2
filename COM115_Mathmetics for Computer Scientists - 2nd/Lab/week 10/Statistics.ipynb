## 8.1. Introduction to Statistics

Statistics is a branch of mathematics concerned with collecting, organizing, analyzing, interpreting, and presenting data. It plays a really important role in different fields such as science, business, economics, or social sciences. It is possible to distinguish two different types of statistics: descriptive statistics and inferential statistics. 

### Descriptive Statistics

Descriptive statistics involves methods for summarizing and describing the characteristics of the data. It focuses on presenting and organizing data in a meaningful way, typically through measures of central tendency (e.g., mean, median, mode) and measures of variability (e.g., range, variance, standard deviation). Descriptive statistics provide insights into the features of the data, helping us understand its distribution and patterns. This lecture will cover the introduction to these topics.

### Inferential Statistics

In contrast, inferential statistics involves getting conclusions about a population based on a sample of data. It extends the findings from a sample to the larger population, allowing us to make predictions, and test hypotheses. Inferential statistics relies on probability theory and statistical methods to estimate parameters, assess uncertainty, and evaluate the significance of relationships within the data. We will be covering some of these topics in the following weeks.

## 8.3. Histograms

A histogram is a graphical representation of the distribution of numerical data. Its purpose is to provide a visual summary of the frequency distribution of a dataset, enabling us to understand the underlying pattern and characteristics of the data at a glance.

In a histogram, the range of values is divided into intervals called bins, and the height of each bar represents the frequency or count of observations within that bin. By visualizing the distribution of data, histograms help in identifying central tendencies, spreads, and outliers within a dataset.

Let us generate a histogram for the values in our previous example:


```python
import matplotlib.pyplot as plt

# Values
values = [1, 5, 8, 4, 6, 3, 1, 7, 5, 4, 3, 2, 1, 5, 6, 7, 8, 9, 3, 4]

# Create histogram. Difine number of bins and colours.
plt.hist(values, bins=12, color='skyblue', edgecolor='black')

# Add labels and title
plt.title('Histogram of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

Interpreting histograms involves understanding the shape, spread, and central tendency of the data distribution. The shape of a histogram can provide insights into the underlying pattern of the data. A symmetric distribution, something like a bell-shaped curve, indicates that the data is evenly distributed around the mean. Skewed distributions, on the other hand, have tails that extend towards one side, suggesting asymmetry in the data.

Additionally, the spread of the histogram reflects the variability or dispersion of the data. A wider spread indicates greater variability, while a narrower spread suggests less variability. The highest point on the histogram typically corresponds to the mode, providing insight into the most frequently occurring values in the dataset.

## 8.4. Measures of Central Tendency

Measures of central tendency are statistical metrics provide different information about the typical value of a dataset. They provide insights into the central or most representative value around which the data is distributed.

### Mean, Median, and Mode
Let us consider a set of values $A=\{x_1,x_2,...,x_n\}$.

- **Mean**: The mean of $A$, also known as the average, is computed by summing up all values in the dataset and dividing this number by the total number of values. 
  
  $$\bar{A} = \frac{1}{n}\sum_{i=1}^{n} x_i$$

- **Median**: The median is the middle value of a dataset when it is ordered from smallest to largest. If there is an even number of values, the median is the average of the two middle values.
  
  Median formula (odd number of values): $$M(A) = x_{\frac{n+1}{2}}$$
  
  Median formula (even number of values): $$M(A) = \frac{1}{2}(x_{\frac{n}{2}} + x_{\frac{n}{2} + 1})$$

- **Mode**: The mode is the value that appears most frequently in the dataset. It can be used for categorical and numerical data, and a dataset may have one mode (unimodal), two modes (bimodal), or more.


Let us recover the set of values from our previous examples. 

**Values:** 

$A=\{1, 5, 8, 4, 6, 3, 1, 7, 5, 4, 3, 2, 1, 5, 6, 7, 8, 9, 3, 4\}$.

Let us compute the mean:
$$\bar{A} = \frac{1}{|A|}\sum_{i=1}^{|A|} x_i=\frac{1}{|20|}(1+ 5+ 8+ 4+ 6+ 3+ 1+ 7+ 5+ 4+ 3+ 2+ 1+ 5+ 6+ 7+ 8+ 9+ 3+ 4)=5,25$$

For the median, we first need to order the dataset. 

$B=ordered(A)=\{1, 1, 1, 2, 3, 3, 3, 4, 4, 4, 5, 5, 5, 6, 6, 7, 7, 8, 8, 9\}$. As we have an even number of observations

$$M(A) = \frac{1}{2}(x_{\frac{|B|}{2}} + x_{\frac{|B|}{2} + 1})=(5+5)/2=5.$$

Finally, to compute the mode, we just need to check our frequency table from $8.2.$, to see that we have a multimodal set where 3, 4, and 5 appear three times each. Therefore:

$$Mo(A)=\{3,4,5\}$$

## 8.5. Dispersion Measures


Dispersion Measures provide insights into the spread of the data, indicating how much the values differ from the central tendency measures. Here are key aspects of measures of variability:


#### Range

The range of a dataset A is the difference between the maximum and minimum values in the dataset.

  $$R(A) = max(A) - min(A)$$

#### Variance and Standard Deviation

Variance measures the average squared deviation of each data point from the mean. It is usually noted by $Var(A)$ or $\sigma^2(A)$.

Standard deviation is the square root of the variance and measures the average deviation of each data point from the mean. It is usually noted by $\sigma(A)$.

$$Var(A)= \sigma^2(A) = \frac{1}{|A|}\sum_{i=1}^{|A|} (x_i - \bar{x})^2$$

$$\sigma(A)=\sqrt{Var(A)}$$

Larger variance and standard deviation values indicate greater variability or dispersion in the dataset. Smaller variance and standard deviation values indicate less variability or dispersion.

#### Quantiles

Quantiles are statistical measures used to divide a dataset into equal-sized intervals or subsets. They provide information about the distribution of the data by dividing it into ordered parts. Commonly used quantiles include percentiles, quartiles, and the median.



**Percentiles**

Percentiles divide the dataset into hundred equal-sized intervals, representing the percentage of data points below a given value.

  - The k-th percentile, denoted as $P_k$, is the value below which $k$% of the data fall.
  - For example, the 75th percentile ($P_{75}$) represents the value below which 75% of the data fall.

**Quartiles**

Quartiles divide the dataset into four equal parts, representing the 25th, 50th, and 75th percentiles.

  - The first quartile ($Q1$) is the value below which 25% of the data fall.
  - The second quartile ($Q2$) is the median, representing the value below which 50% of the data fall.
  - The third quartile ($Q3$) is the value below which 75% of the data fall.

**Interquartile Range (IQR)**

The interquartile range (IQR) is the range of values between the first and third quartiles.

  $$IQR(A) = Q3(A) - Q1(A)$$

## 8.6. Now using python

```python
import numpy as np
from scipy import stats

# Given dataset
data = [1, 5, 8, 4, 6, 3, 1, 7, 5, 4, 3, 2, 1, 5, 6, 7, 8, 9, 3, 4]

# Mean
mean = np.mean(data)

# Median
median = np.median(data)

# Range
data_range = np.max(data) - np.min(data)

# Variance
variance = np.var(data)

# Standard deviation
std_dev = np.std(data)

# Percentile 43
percentile_43 = np.percentile(data, 43)

# Quartiles
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)

# Interquartile range (IQR)
iqr = q3 - q1

# Print results
print("Mean:", mean)
print("Median:", median)

print("Range:", data_range)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Percentile 43:", percentile_43)
print("Q1 (25th percentile):", q1)
print("Q3 (75th percentile):", q3)
print("Interquartile Range (IQR):", iqr)
```

## 8.7. Univariate vs. Bivariate (Correlation metrics)

Univariate analysis focuses on examining one variable at a time, exploring its distribution, central tendency, and variability. It helps in understanding the characteristics and patterns within a single variable.

Bivariate analysis, on the other hand, involves the simultaneous examination of two variables to understand the relationship or association between them. It aims to explore how changes in one variable are associated with changes in another variable. This analysis is valuable for discovering dependencies, patterns, and trends between pairs of variables.


#### Covariance

Covariance of X and Y measures the joint variability of two random variables. It indicates the degree to which two variables change together. If the covariance between two variables is positive, it suggests that they tend to increase or decrease together. Conversely, if the covariance is negative, it suggests that they tend to change in opposite directions.

**Formula**

The covariance between two variables X and Y is calculated as follows:

$$
\text{cov}(X, Y) = \frac{\sum_{i=1}^n (X_i - \bar{X})(Y_i - \bar{Y})}{n}
$$

Where:
- $X_i$ and $Y_i$ are individual data points,
- $\bar{X}$ and $\bar{Y}$ are the means of X and Y respectively, and
- $n$ is the number of data points.



#### Correlation Metrics

Correlation metrics quantify the degree and direction of the relationship between two variables in a dataset. Most commonly used correlation coefficient is the Pearson's correlation.

**Pearson Correlation Coefficient**
The Pearson correlation coefficient measures the linear relationship between two continuous variables. It ranges from -1 to +1, where:
- +1 indicates a perfect positive linear relationship,
- 0 indicates no linear relationship, and
- -1 indicates a perfect negative linear relationship.

The formula to calculate the Pearson correlation coefficient between variables X and Y is:

$$ r = \frac{\sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum_{i=1}^{n}(X_i - \bar{X})^2 \sum_{i=1}^{n}(Y_i - \bar{Y})^2}} $$

Interpretation of the Pearson correlation coefficient:
- Values close to +1 or -1 indicate a strong linear relationship.
- Values close to 0 indicate a weak or no linear relationship.
- Positive values imply a positive association, while negative values imply a negative association.

Let us do a quick example. Consider the following two variables:

| X  | 3  | 5  | 2  | 9  | 11 | 7  | 8  | 6  | 12 | 4  |
|----|----|----|----|----|----|----|----|----|----|----|
| Y  | 7  | 11 | 5  | 19 | 23 | 15 | 17 | 13 | 25 | 9  |



```python
import numpy as np

# Given data
X = [3, 5, 2, 9, 11, 7, 8, 6, 12, 4]
Y = [7, 11, 5, 19, 23, 15, 17, 13, 25, 9]

# Compute Pearson correlation coefficient
pearson_coefficient = np.corrcoef(X, Y)[-1, 1]

pearson_coefficient
```

As you can see, we are getting a Pearsons' Coefficient of $r=0.9999$, meaning that both variables have a strong linear relationship. What does this mean? 

Basically, it means that one variable is a linear transformation of the other, that is, $Y=a*X+b$, where $a,b\in \mathbb{R}$. Let us draw this:

```python
import matplotlib.pyplot as plt

# Given data
X = [3, 5, 2, 9, 11, 7, 8, 6, 12, 4]
Y = [7, 11, 5, 19, 23, 15, 17, 13, 25, 9]

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Data Points')
plt.title('Scatter Plot of Variables X and Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

## 8.8. Linear Regression


Simple linear regression is a statistical method used to model the relationship between two variables: one independent variable (often denoted as $X$) and one dependent variable (often denoted as $Y$). It assumes that there is a linear relationship between the independent variable and the dependent variable (you could use Pearson's coef. to check this).

**Equation**

The relationship between the independent variable $X$ and the dependent variable $Y$ is modeled using a straight line equation:

$$ Y = \beta_0 + \beta_1 X + \varepsilon $$

Where:
- $Y$ is the dependent variable,
- $X$ is the independent variable,
- $\beta_0$ is the intercept of the line (the value of $Y$ when $X$ is 0),
- $\beta_1$ is the slope of the line (the change in $Y$ for a one-unit change in $X$), and
- $\varepsilon$ is the error term (the difference between the observed and predicted values of $Y$).

**Goal**

The goal of simple linear regression is to estimate the values of $\beta_0$ and $\beta_1$ that minimize the sum of squared errors (the difference between the observed and predicted values of $Y$).

**Usage**

Simple linear regression is commonly used to:
- Understand the relationship between two variables,
- Make predictions about the dependent variable based on the independent variable, and
- Assess the strength and direction of the relationship between the variables.

Let us work on a made up example:


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate random data for demonstration
np.random.seed(0)
X = 2 * np.random.rand(100, 1)   # Independent variable
Y = 4 + 3 * X + np.random.randn(100, 1)   # Dependent variable

# Plot the data
plt.scatter(X, Y, color='blue')
plt.title('Scatter Plot of X and Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

# Fit the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Get the model parameters
intercept = model.intercept_[0]
slope = model.coef_[0][0]

# Print the model parameters
print("Intercept (beta0):", intercept)
print("Slope (beta1):", slope)

# Plot the regression line
plt.scatter(X, Y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title('Linear Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
