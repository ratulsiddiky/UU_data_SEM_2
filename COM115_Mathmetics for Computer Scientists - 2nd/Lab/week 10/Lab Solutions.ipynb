## Exercise 1 



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Jorge's dice throws
throws = [6, 2, 6, 6, 3, 6, 1, 6, 4, 3, 2, 6, 1, 6, 5, 4, 6, 6, 3, 4]

# Mean
mean = np.mean(throws)
print("Mean of the outcomes:", mean)

# Variance
variance = np.var(throws)
print("Variance of the outcomes:", variance)


# Standard Deviation
std_dev = np.std(throws)
print("Standard Deviation of the outcomes:", std_dev)

# Histogram
plt.hist(throws, bins=range(1, 8), align='left', rwidth=0.8, edgecolor='black')
plt.title('Histogram of Dice Throws')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.xticks(range(1, 7))
plt.grid(axis='y', alpha=0.5)
plt.show()
```

## Exercise 2 



```python
import numpy as np
import matplotlib.pyplot as plt

# Given list of incomes
incomes = [25000, 18000, 32000, 27000, 35000, 28000, 20000, 38000, 23000, 29000,
           27000, 33000, 24000, 19000, 32000, 36000, 26000, 22000, 31000, 27000]

# Convert the list to numpy array for easier manipulation
incomes = np.array(incomes)

# Descriptive statistics before the rich person joined
mean_salary_before = np.mean(incomes)
most_common_salary_before = np.bincount(incomes).argmax()
variance_before = np.var(incomes)
q1_before = np.percentile(incomes, 25)

# Print results before the rich person joined
print("Before the rich person joined:")
print("Mean salary: £{:.2f}".format(mean_salary_before))
print("Most common salary: £{}".format(most_common_salary_before))
print("Variance: £{:.2f}".format(variance_before))
print("Quartile 1 (Q1): £{:.2f}".format(q1_before))

# Plot histogram before the rich person joined
plt.hist(incomes, bins=10, edgecolor='black')
plt.title('Income Distribution in the Neighborhood (Before)')
plt.xlabel('Income (£)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# Exercise 3

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
data = {
    'Size (sq. ft.)': [2453, 1635, 1563, 2531, 1833, 1077, 2578, 2628, 1505, 2935,
                       2501, 1337, 2920, 2740, 2963, 1776, 1555, 2846, 2671, 899,
                       2808, 1555, 1597, 1459, 1223, 1439, 1344, 1514, 951, 2007],
    'Price (pounds)': [243078.84, 161774.07, 151891.00, 250886.73, 180595.42,
                       101083.86, 257236.01, 267336.73, 154576.35, 294645.49,
                       244969.11, 136087.63, 298463.49, 270342.71, 288272.99,
                       182547.38, 156054.07, 282695.34, 267674.80, 91626.56,
                       272125.21, 163791.76, 171198.86, 143544.32, 128613.58,
                       138047.42, 139728.95, 147900.31, 95820.40, 202692.71]
}

# Create DataFrame
df = pd.DataFrame(data)

# 1. Data Exploration
plt.figure(figsize=(10, 6))
plt.scatter(df['Size (sq. ft.)'], df['Price (pounds)'], color='blue')
plt.title('House Prices vs House Sizes')
plt.xlabel('Size (sq. ft.)')
plt.ylabel('Price (pounds)')
plt.grid(True)
plt.show()

# 2. Linear Regression Analysis
X = df[['Size (sq. ft.)']]
y = df['Price (pounds)']

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Coefficients
slope = model.coef_[0]
intercept = model.intercept_
print("Regression Coefficients:")
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")

# Coefficient of determination (R-squared)
r_squared = model.score(X, y)
print(f"\nCoefficient of Determination (R-squared): {r_squared:.2f}")

# Plot the regression line
plt.figure(figsize=(10, 6))
plt.scatter(df['Size (sq. ft.)'], df['Price (pounds)'], color='blue')
plt.plot(X, model.predict(X), color='red', linewidth=2)
plt.title('House Prices vs House Sizes with Linear Regression')
plt.xlabel('Size (sq. ft.)')
plt.ylabel('Price (pounds)')
plt.grid(True)
plt.show()

# 3. Prediction
size_prediction = np.array([[2500]])
price_prediction = model.predict(size_prediction)
print(f"\nPredicted price for a house with size 2500 sq. ft.: {price_prediction[0]:,.2f} pounds")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame from the provided data
data = {
    'Person': range(1, 31),
    'Income (pounds)': [2500, 3200, 2800, 3500, 3100, 2700, 3300, 2900, 2600, 3000, 
                        3200, 2800, 3100, 2900, 2700, 3300, 3400, 3100, 2900, 3000, 
                        2800, 3200, 3300, 3100, 2800, 3000, 3100, 2900, 2700, 3300],
    'Expenditure (pounds)': [1800, 2100, 1900, 2200, 2000, 1900, 2100, 2000, 1800, 2200, 
                              2300, 2100, 2000, 1900, 1800, 2200, 2400, 2100, 2000, 2200, 
                              1900, 2100, 2200, 2000, 1800, 2200, 2300, 2000, 1900, 2200]
}
df = pd.DataFrame(data)

# Scatter plot of income against expenditure
plt.figure(figsize=(10, 6))
plt.scatter(df['Income (pounds)'], df['Expenditure (pounds)'], color='blue', alpha=0.7)
plt.title('Scatter Plot of Income vs. Expenditure')
plt.xlabel('Income (pounds)')
plt.ylabel('Expenditure (pounds)')
plt.grid(True)
plt.show()
# Compute covariance
covariance = df['Income (pounds)'].cov(df['Expenditure (pounds)'])
print(f'Covariance between income and expenditure: {covariance:.2f}')

# Calculate variance and standard deviation
income_variance = df['Income (pounds)'].var()
income_std = df['Income (pounds)'].std()
expenditure_variance = df['Expenditure (pounds)'].var()
expenditure_std = df['Expenditure (pounds)'].std()

print(f'Income variance: {income_variance:.2f}, Income standard deviation: {income_std:.2f}')
print(f'Expenditure variance: {expenditure_variance:.2f}, Expenditure standard deviation: {expenditure_std:.2f}')


# Calculate Pearson correlation coefficient
correlation_coefficient = df['Income (pounds)'].corr(df['Expenditure (pounds)'])
print(f'Pearson correlation coefficient: {correlation_coefficient:.2f}')

from scipy.stats import linregress

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(df['Income (pounds)'], df['Expenditure (pounds)'])

print(f'Regression coefficients: Slope={slope:.2f}, Intercept={intercept:.2f}')
print(f'Coefficient of determination (R-squared): {r_value**2:.2f}')


# Scatter plot of income against expenditure
plt.figure(figsize=(10, 6))
plt.scatter(df['Income (pounds)'], df['Expenditure (pounds)'], color='blue', alpha=0.7, label='Data Points')

# Plot linear regression line
x_values = np.linspace(df['Income (pounds)'].min(), df['Income (pounds)'].max(), 100)
y_values = slope * x_values + intercept
plt.plot(x_values, y_values, color='red', linestyle='--', label='Linear Regression Line')

plt.title('Scatter Plot of Income vs. Expenditure with Linear Regression')
plt.xlabel('Income (pounds)')
plt.ylabel('Expenditure (pounds)')
plt.legend()
plt.grid(True)
plt.show()
```
