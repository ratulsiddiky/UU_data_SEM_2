## Quadratic Equations

As discussed in section 1.9 of the lecture notes, quadratic equations can be expressed in the form

$ax^2 + bx + c = 0$

where $a$, $b$ and $c$ are real numbers. The solutions, i.e. the values of $x$ that solve the equation are obtained from the formula

$x = \frac{-b \pm \sqrt{b^2-4ac}}{2a}$.

In other words there are two solutions

$x = \frac{-b + \sqrt{b^2-4ac}}{2a}$ and $x = \frac{-b - \sqrt{b^2-4ac}}{2a}$.

Example 1.10.1 in the notes shows how to work out the solutions for the equation

$2x^2 + 7x - 4 = 0$

Since $a = 2$, $b = 7$ and $c = -4$ in this case, these values can be put into the formula to generate the solutions $x = -4$ and $x = 0.5$ (see example 1.10.1 for details).

The following code shows how to calculate the answers in Python.
(Note: this code would not work if $b^2 < 4ac$ since it would involve taking the square root of a negative number which we won't consider here.)

```python
import numpy as np
a = 2; b = 7; c = -4
x1=x2=0
# First solution
x1 = (-b + np.sqrt(b**2 -4*a*c))/(2*a)
print(x1)

# Second solution
x2 = (-b - np.sqrt(b**2 -4*a*c))/(2*a)
print(x2)
```

## Plotting Data Using Matplotlib
<a href="https://matplotlib.org/">Matplotlib</a> is a very useful library for producing figures in Python. (We will also come across another plotting library called *Seaborn* which is based on Matplotlib.) Here we will use the *Pyplot* interface within Matplotlib. For further details, see https://matplotlib.org/stable/tutorials/pyplot.html You should look at these resources to explore more details about how to plot and edit figures.

Here is a simple example to illustrate how to produce a x-y line plot.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,1,2,3,4,5])
y = np.array([1,2,5,10,17,26])

plt.plot(x, y) 
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

## Plotting Functions Using Matplotlib

To plot a mathematical function we can use the Numpy function `linspace` to create an array of $x$ values within a specified range and then use the equation to determine the values of another array $y$. The example below plots the function $y = 3x - 2$ for $x$ values from 0 to 5. Here 100 $x$ values are used. (Note: a legend is included to label the plot. This becomes more relevant when there are multiple plots. See exercise 3.)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,5,100)
y = 3 * x - 2
plt.plot(x, y, label='y=3x-2')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.show()
```
